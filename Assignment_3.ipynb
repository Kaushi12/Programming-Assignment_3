```python
num1=int(input("Enter any number: "))
if num1>0:
    print("Positive Number")
elif num1<0:
    print("Negative Number")
else:
    print("Zero")
```

```python
num2=int(input("Enter any number: "))
if (num2%2)==0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
year=int(input("Enter a year: "))
if(year%400)==0:
    print("Leap year")
elif(year%4)==0:
    print("Leap year")
elif(year%100)==0:
    print("Not a leap year")
```

```python
num3=int(input("Enter any number: "))
i=2
while (i!=num3):
    if (num3%i)==0:
        print("Not a Prime number")
        i+=1
    else:
        print("Prime Number")
    break
        
    
    
```

```python
num4=int(input("Enter any number: "))
if num4>1:
    for i in range(2,num4):
        if (num4%i)==0:
            print("Not a Prime number")
            i+=1
        else:
            print("Prime Number")
        break
            
            
else:
    print("Not a Prime Number")

        
        
```
